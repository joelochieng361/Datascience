```python
import pandas
data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
#Use the describe function
data.describe()
```

```python
#We have noticed that the itemprice column contains three empty slots, We need to fill this empty slots with a value
#We can use the mean of the column to fill the empty slots
data["itemprice"] = data["itemprice"].fillna(55)

#describe the data again to see the changes
data.describe()
```

```python
# get the number of entries in a given column
data.groupby("gender").size()
```

```python
#combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
#Import the modules
import seaborn as sns #for data visualization and statistical graphics
import matplotlib.pyplot as plt #for data visualization and plotting
```

```python
# Below is a count plot showing the distriution of males to females
sns.countplot(x="gender", data=data, palette="autumn")
plt.title("Distribution of Males to Females")
plt.ylabel("Total count of individuals")
plt.xlabel("Gender")
```

```python
#create a count plot for products
#which product category appears most
sns.countplot(x="product", data=data, palette="autumn")
plt.title("Distribution of Product Categories")
plt.ylabel("Total count of products")
plt.xticks(rotation=45)
```

```python
#Below is a count that has two columns
sns.countplot(x="gender", hue="city", palette="Reds", data=data)
```

```python
# below is a pieplot
data["payment"].value_counts().plot.pie(autopct='%1.1f%%')
```
